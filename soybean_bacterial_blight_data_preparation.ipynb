```python
# import library yang dibutuhkan

import os
import shutil # operasi file
from sklearn.model_selection import train_test_split # splitting dataset
from glob import glob # untuk cari file dengan pola tertentu
```

## Prepare Folder yang Dibutuhkan

```python
# prepare folder dataset

DATA_DIR = "/content/drive/MyDrive/Dataset/soybean_bacterial_blight_dataset/soybean_bacterial_blight" # path to dataset
os.makedirs(f"{DATA_DIR}/train/images", exist_ok=True)
os.makedirs(f"{DATA_DIR}/train/masks", exist_ok=True)
os.makedirs(f"{DATA_DIR}/val/images", exist_ok=True)
os.makedirs(f"{DATA_DIR}/val/masks", exist_ok=True)
```

## Get Images

```python
# get original images (tanpa _black dan _label)

all_images = glob(f"{DATA_DIR}/*.jpg") + glob(f"{DATA_DIR}/*.png") # get all images
original_images = [f for f in all_images if "_black" not in f and "_label" not in f] # filter _black dan _label
```

```python
print("Jumlah file original:", len(original_images))
```

## Split Testing dan Validation

```python
# split train dan val

train_imgs, val_imgs = train_test_split(original_images, test_size=0.2, random_state=42) # split data train dan val dengan ukuran 80:20
```

```python
print("Jumlah file train:", len(train_imgs))
print("Jumlah file val:", len(val_imgs))
```

```python
# function copy files
def copy_files(file_list, split):
  for img_path in file_list:
    base_name = os.path.splitext(os.path.basename(img_path))[0] # ambil nama file
    ext = os.path.splitext(img_path)[1] # ambil extension

    # copy original image
    shutil.copy(img_path, f"{DATA_DIR}/{split}/images/{base_name}{ext}")

    # copy label mask
    label_path = img_path.replace(f"{base_name}{ext}", f"{base_name}_label.png")
    # mencari file dengan _label dan mencopy ke dalam /masks
    if os.path.exists(label_path):
      shutil.copy(label_path, f"{DATA_DIR}/{split}/masks/{base_name}.png")
```

```python
# task: copy file
copy_files(train_imgs, "train")
copy_files(val_imgs, "val")
```

```python
print("Jumlah file train:", len(train_imgs))
print("Jumlah file val:", len(val_imgs))
```
